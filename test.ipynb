```python
import geostatspy.geostats as geostats
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
rawdata = pd.read_csv('./data.csv')
```

```python
tmin=-1.0e21
tmax=1.0e21
lag=5
lag_tol=2.5
nlag=60
azi=0
azi_tol=90
bandwidth=200
```

```python
rawdata['NAg'], _, _ = geostats.nscore(rawdata, 'Ag')
```

```python
np.var(rawdata['Au'])
```

```python
for z in rawdata.columns[2:]:
    print(z)
    # rawdata['N'+ z], _, _ = geostats.nscore(rawdata, z)
    lags, gammas, npps = geostats.gamv(rawdata,"locx","locy", z,tmin,tmax,lag,lag_tol,nlag,azi,azi_tol,bandwidth,isill=1.0)
    plt.figure()
    plt.scatter(lags,gammas,color = 'darkorange',edgecolor='black',s = 0.05,label = 'Azimuth ' +str(azi))
    plt.plot([0,200],[np.var(rawdata[z]),np.var(rawdata[z])],color = 'black')
    plt.title(r'variogram-'+z)
    plt.xlabel(r'Lag Distance $\bf(h)$, (m)')
    plt.ylabel(r'$\gamma \bf(h)$')
    plt.savefig('./result/'+z+'.png')
```
